```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve,auc
```

```python
f1 = open(".csv","rb")
target =np.loadtxt(f1,delimiter=',',skiprows=0)
f1.close()
target = np.array(target)
print(target.shape)
```

```python
f2 = open(".csv","rb")
features2 =np.loadtxt(f2,delimiter=',',skiprows=0)
f2.close()
Mismatchy_proba= np.array(features2)
Mismatchy_proba=Mismatchy_proba[:,1]
f3 = open(".csv","rb")
features3 =np.loadtxt(f3,delimiter=',',skiprows=0)
f3.close()
DACy_proba= np.array(features3)
DACy_proba=DACy_proba[:,1]
f4 = open(".csv","rb")
features4 =np.loadtxt(f4,delimiter=',',skiprows=0)
f4.close()
DCCy_proba= np.array(features4)
DCCy_proba=DCCy_proba[:,1]
f5 = open(".csv","rb")
features5 =np.loadtxt(f5,delimiter=',',skiprows=0)
f5.close()
DSAy_proba= np.array(features5)
DSAy_proba=DSAy_proba[:,1]
f6 = open(".csv","rb")
features6 =np.loadtxt(f6,delimiter=',',skiprows=0)
f6.close()
DSET= np.array(features6)
```

```python
fpr1,tpr1,threshold1 = roc_curve(target,Mismatchy_proba)
roc_auc1 = auc(fpr1,tpr1)
fpr2,tpr2,threshold2 = roc_curve(target,DACy_proba)
roc_auc2 = auc(fpr2,tpr2)
fpr3,tpr3,threshold3 = roc_curve(target,DCCy_proba)
roc_auc3 = auc(fpr3,tpr3)
fpr4,tpr4,threshold4 = roc_curve(target,DSAy_proba)
roc_auc4 = auc(fpr4,tpr4)
fpr5,tpr5,threshold5 = roc_curve(target,DSET)
roc_auc5 = auc(fpr5,tpr5)
```

```python
plt.figure()
plt.figure(figsize=(7,6))
plt.plot(fpr1,tpr1,color = 'orange',linewidth=1,label='Mismatch ROC curve(AUC= %0.4f)'%roc_auc1)
plt.plot(fpr2,tpr2,color = 'green',linewidth=1,label='DAC ROC curve (AUC= %0.4f)'%roc_auc2)
plt.plot(fpr3,tpr3,color = 'blue',linewidth=1,label='DCC ROC curve (AUC= %0.4f)'%roc_auc3)
plt.plot(fpr4,tpr4,color = 'pink',linewidth=1,label='DSA ROC curve (AUC= %0.4f)'%roc_auc4)
plt.plot(fpr5,tpr5,color = 'red',linewidth=1,label='WDSET ROC curve (AUC= %0.4f)'%roc_auc5)
plt.plot([0,1],[0,1],color = 'black',linewidth=1,linestyle='-.')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel("1-Specificity")
plt.ylabel("Sensitivity")
#plt.title('The ROC curve for identifying enhancers and non enhancers.')
plt.legend(loc="lower right")
plt.show()
```
